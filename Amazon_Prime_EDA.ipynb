## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# While Checking the duplicates, already drop the duplictes.
titles['release_year'] = pd.to_numeric(titles['release_year'], errors='coerce')
titles['imdb_score'] = pd.to_numeric(titles['imdb_score'], errors='coerce')
titles['imdb_votes'] = pd.to_numeric(titles['imdb_votes'], errors='coerce')
titles['tmdb_popularity'] = pd.to_numeric(titles['tmdb_popularity'], errors='coerce')
titles['tmdb_score'] = pd.to_numeric(titles['tmdb_score'], errors='coerce')

# Filling Imdb Score with mean
titles['imdb_score'].fillna(titles['imdb_score'].mean(),inplace =True)
print(f'Null Value in imdb score -',titles['imdb_score'].isnull().sum())
#Filling Null value in votes
titles['tmdb_score'].fillna(titles['tmdb_score'].median(),inplace =True)
print(f'Null Value in tmdb score -',titles['tmdb_score'].isnull().sum())
````

````python
titles['type']= titles['type'].replace({'MOVIE':'Movie', 'SHOW':'TV Shows'})
titles
````
